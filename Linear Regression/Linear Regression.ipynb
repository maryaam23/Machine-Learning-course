### Apply the hypothesis function on make regression dataset 

```python
# Generating dataset
from sklearn.model_selection import train_test_split
X,y =make_regression(n_features=3,n_samples=500,random_state=10)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.40)
```

```python
# hypothesis function
def hypothesis(theta0,theta1,theta2,theta3,x1,x2,x3):
    y_hat = (3*theta1*(x1**2)) + (theta2*(x2**3)) - (theta3*(x2**3)) + (theta2*theta3*(x3**3)) + theta0
    return y_hat
```

```python
# Fit Function 
def poly(y_train,lr=0.001):
    
    theta0 = 1
    theta1 = 1.9
    theta2 = -1.5
    theta3 = -12
    
    
    losses = []
    for itr in range(3000):
        diffs = []
        grad0 = []
        grad1 = []
        grad2 = []
        grad3 = []
        
        for i in range(X_train.shape[0]):

            
            y_hat = hypothesis(theta0,theta1,theta2,theta3,X_train[i,0],X_train[i,1],X_train[i,2])
            
            diff  = (y_hat - y_train[i])
            diff2 = diff**2
            diffs.append(diff2)
            
            g0 = diff 
            grad0.append(g0)
            
            g1 = diff * 3 * (X_train[i,0])**2
            grad1.append(g1)
            
            g2 = diff * ((X_train[i,1])**3 + (theta3 * (X_train[i,2])**3))
            grad2.append(g2)
            
            g3 = diff * (((-1)*(X_train[i,1])**3) + (theta2 * (X_train[i,2])**3))
            grad3.append(g3)
            
        
        
        loss = (1/(2*X_train.shape[0])) * sum(diffs)
        losses.append(loss)
        
        theta0 = theta0 - (lr/X_train.shape[0]) * sum(grad0)
        theta1 = theta1 - (lr/X_train.shape[0]) * sum(grad1)
        theta2 = theta2 - (lr/X_train.shape[0]) * sum(grad2)
        theta3 = theta3 - (lr/X_train.shape[0]) * sum(grad3)
        
    return theta0,theta1,theta2,theta3,losses
            
```

```python
th0,th1,th2,th3,losses = poly(y_train)
```

```python
print(th0,th1,th2,th3)
```

```python
plt.figure()
plt.scatter(x=list(range(0, 3000)), y=losses)
plt.title('Losses Graph')
plt.show()
```

```python
# Plot real values against y predicted
y_preds = []
for i in range(X_test.shape[0]):
    y_pred = hypothesis(th0,th1,th2,th3,X_test[i,0],X_test[i,1],X_test[i,2])
    y_preds.append(y_pred)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x=range(200),y= y_test)           
plt.scatter(x=range(200), y=y_preds)
plt.legend(['y_test','y_pred'],loc='lower right');
```

## a.

```python
import numpy as np
from sklearn.datasets import make_regression

X,y =make_regression(n_features=15,n_samples=500,random_state=10)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

## b.

# Linear regression

```python
from sklearn.linear_model import LinearRegression

lin_model = LinearRegression()
```

```python
lin_model.fit(X_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error
y_predicted_lin = lin_model.predict(X_test)
mean_error_lin = mean_squared_error(y_test,y_predicted_lin)
print(mean_error_lin)
```

## Lasso regression

```python
from sklearn import linear_model
lasso_model = linear_model.Lasso()
lasso_model.fit(X_train,y_train)
```

```python
y_predicted_lasso = lasso_model.predict(X_test)
mean_error_lasso = mean_squared_error(y_test,y_predicted_lasso)
print(mean_error_lasso)
```

# Ridge regression 

```python
from sklearn.linear_model import Ridge
ridge_model = Ridge()
ridge_model.fit(X_train,y_train)
```

```python
y_predicted_ridge = ridge_model.predict(X_test)
mean_error_ridge = mean_squared_error(y_test,y_predicted_ridge)
print(mean_error_ridge)
```

# Polynomial linear regression 

```python
from sklearn.preprocessing import PolynomialFeatures
poly_model = PolynomialFeatures(2)
X_train_transform = poly_model.fit_transform(X_train)
```

```python
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_train_transform, y_train)
```

```python
X_test_transform = poly_model.fit_transform(X_test)
```

```python
y_predicted_poly = lin_reg_2.predict(X_test_transform)
mean_error_poly = mean_squared_error(y_test,y_predicted_poly)
print(mean_error_poly)
```
